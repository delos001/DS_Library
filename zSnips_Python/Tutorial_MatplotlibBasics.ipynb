## MatplotLib Crash Course

#### Line Plot

```python
## simple line from 1-D data

myarray = numpy.array([1, 2, 3])

plt.plot(myarray)
plt.xlabel('X-axis Name Here')
plt.ylabel('Y-axis Name Here')
plt.show()
```

#### Scatter Plot

```python
x = numpy.array([1, 2, 3, 4])
y = numpy.array([3, 4, 5, 6])

plt.scatter(x, y)
plt.xlabel('X-axis Name Here')
plt.ylabel('Y-axis Name Here')
plt.show()
```
